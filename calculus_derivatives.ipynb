```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

print("="*60)
print("📐 CALCULUS: DERIVATIVES")
print("="*60)

# ========== 1. SYMBOLIC DIFFERENTIATION ==========
print("\n1. SYMBOLIC DIFFERENTIATION")
print("-"*40)

x = sp.Symbol('x')

# Polynomial
f1 = x**3 + 2*x**2 - 5*x + 3
df1 = sp.diff(f1, x)

print(f"f(x) = {f1}")
print(f"f'(x) = {df1}")

# Exponential
f2 = sp.exp(2*x)
df2 = sp.diff(f2, x)
print(f"\nf(x) = {f2}")
print(f"f'(x) = {df2}")

# Trigonometric
f3 = sp.sin(3*x)
df3 = sp.diff(f3, x)
print(f"\nf(x) = {f3}")
print(f"f'(x) = {df3}")

# ========== 2. PARTIAL DERIVATIVES ==========
print("\n" + "="*60)
print("2. PARTIAL DERIVATIVES")
print("="*60)

x, y = sp.symbols('x y')

# Function of two variables
f = x**2 * y + 3*x * y**2 + 2*x + y

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print(f"f(x,y) = {f}")
print(f"\n∂f/∂x = {df_dx}")
print(f"∂f/∂y = {df_dy}")

# Gradient (vector of partial derivatives)
gradient = sp.Matrix([df_dx, df_dy])
print(f"\nGradient = {gradient}")

# ========== 3. CHAIN RULE ==========
print("\n" + "="*60)
print("3. CHAIN RULE")
print("="*60)

# f(x) = sin(x^2 + 3x)
inner = x**2 + 3*x
outer = sp.sin(inner)
derivative = sp.diff(outer, x)

print(f"f(x) = sin(x² + 3x)")
print(f"f'(x) = {derivative}")

# With multiple variables
x, y = sp.symbols('x y')
z = x**2 + y**2
f = sp.sin(z)

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print(f"\nz = x² + y²")
print(f"f = sin(z)")
print(f"\n∂f/∂x = {df_dx}")
print(f"∂f/∂y = {df_dy}")

# ========== 4. GRADIENT DESCENT (THE CORE OF AI) ==========
print("\n" + "="*60)
print("4. GRADIENT DESCENT")
print("="*60)

# Function to minimize: f(x) = x^2 + 3x + 1
def f(x):
    return x**2 + 3*x + 1

def df(x):
    return 2*x + 3

# Gradient descent
def gradient_descent(start_x, learning_rate, iterations):
    x_path = [start_x]
    for _ in range(iterations):
        x_current = x_path[-1]
        gradient = df(x_current)
        x_next = x_current - learning_rate * gradient
        x_path.append(x_next)
    return x_path

# Try different learning rates
start_x = 10
iterations = 20

print(f"Finding minimum of f(x) = x² + 3x + 1")
print(f"Starting at x = {start_x}")

# Learning rate = 0.1
lr1 = 0.1
path1 = gradient_descent(start_x, lr1, iterations)
print(f"\nLearning rate = {lr1}:")
print(f"Final x = {path1[-1]:.4f}")

# Learning rate = 0.5
lr2 = 0.5
path2 = gradient_descent(start_x, lr2, iterations)
print(f"\nLearning rate = {lr2}:")
print(f"Final x = {path2[-1]:.4f}")

# ========== 5. VISUALIZE GRADIENT DESCENT ==========
print("\n" + "="*60)
print("5. VISUALIZE GRADIENT DESCENT")
print("="*60)

x_vals = np.linspace(-8, 12, 200)
y_vals = f(x_vals)

plt.figure(figsize=(12, 4))

# Learning rate 0.1
plt.subplot(1, 3, 1)
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x)')
y_path1 = [f(x) for x in path1]
plt.scatter(path1[::2], y_path1[::2], color='red', zorder=5, s=30)
plt.plot(path1, y_path1, 'r--', alpha=0.5)
plt.axvline(x=-1.5, color='green', linestyle='--', alpha=0.5)
plt.title(f'LR = {lr1}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)

# Learning rate 0.5
plt.subplot(1, 3, 2)
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x)')
y_path2 = [f(x) for x in path2]
plt.scatter(path2[::2], y_path2[::2], color='red', zorder=5, s=30)
plt.plot(path2, y_path2, 'r--', alpha=0.5)
plt.axvline(x=-1.5, color='green', linestyle='--', alpha=0.5)
plt.title(f'LR = {lr2}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)

# Convergence comparison
plt.subplot(1, 3, 3)
plt.plot(range(len(path1)), path1, 'b-', label=f'LR={lr1}')
plt.plot(range(len(path2)), path2, 'r-', label=f'LR={lr2}')
plt.axhline(y=-1.5, color='green', linestyle='--', label='Optimal')
plt.title('Convergence Comparison')
plt.xlabel('Iteration')
plt.ylabel('x value')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n🤖 KEY INSIGHT: Learning rate controls convergence speed!")

# ========== 6. GRADIENT DESCENT FOR 2D ==========
print("\n" + "="*60)
print("6. GRADIENT DESCENT IN 2D")
print("="*60)

# Function: f(x,y) = x^2 + y^2 (simple paraboloid)
def f2d(x, y):
    return x**2 + y**2

def df2d(x, y):
    return 2*x, 2*y  # Gradient: (2x, 2y)

# Run gradient descent
x, y = 5, 5  # Start
lr = 0.1
iterations = 20

x_path = [x]
y_path = [y]

for _ in range(iterations):
    grad_x, grad_y = df2d(x, y)
    x = x - lr * grad_x
    y = y - lr * grad_y
    x_path.append(x)
    y_path.append(y)

# Visualize
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create meshgrid
X = np.linspace(-6, 6, 30)
Y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

# Plot surface
ax.plot_surface(X, Y, Z, alpha=0.5, cmap='coolwarm')

# Plot path
z_path = [x**2 + y**2 for x, y in zip(x_path, y_path)]
ax.plot(x_path, y_path, z_path, 'r.-', linewidth=2, markersize=8)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.set_title('Gradient Descent on 2D Paraboloid')

plt.show()

print(f"Started at (5, 5)")
print(f"Ended at ({x_path[-1]:.4f}, {y_path[-1]:.4f})")
print(f"Minimum is at (0, 0)")

print("\n" + "="*60)
print("✅ CALCULUS COMPLETE!")
print("="*60)

print("""
📌 KEY TAKEAWAYS:

1. Derivatives = rate of change
2. Partial derivatives = rate of change for one variable
3. Gradient = vector of all partial derivatives
4. Chain rule = backpropagation in neural networks
5. Gradient descent = main optimization algorithm in AI
6. Learning rate controls how fast we learn

🤖 WHY THIS MATTERS IN AI:
- Neural networks learn by gradient descent
- Backpropagation uses the chain rule
- We minimize loss functions
- Learning rate is a key hyperparameter
""")
```
