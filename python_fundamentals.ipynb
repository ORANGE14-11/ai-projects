```python
# ========== IF STATEMENTS ==========
print("="*50)
print("IF STATEMENTS")
print("="*50)

age = 18
has_license = True

# Basic if
if age >= 18:
    print("You are an adult!")

# if-else
if age >= 18:
    print("You can vote!")
else:
    print("You are too young to vote.")

# if-elif-else
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score}, Grade: {grade}")

# Multiple conditions
if age >= 18 and has_license:
    print("You can drive!")
elif age >= 18 and not has_license:
    print("You need a license to drive.")
else:
    print("You are too young to drive.")

# Comparison operators
print("\nComparison examples:")
x = 10
y = 5

print(f"{x} == {y}: {x == y}")   # Equal
print(f"{x} != {y}: {x != y}")   # Not equal
print(f"{x} > {y}: {x > y}")     # Greater than
print(f"{x} < {y}: {x < y}")     # Less than
print(f"{x} >= {y}: {x >= y}")   # Greater than or equal
print(f"{x} <= {y}: {x <= y}")   # Less than or equal

# ========== EXERCISES ==========
print("\n" + "="*50)
print("YOUR TURN!")
print("="*50)
print("1. Write code to check if a number is even or odd")
print("2. Check if a year is a leap year")
print("3. Grade calculator (A: 90+, B: 80-89, C: 70-79, etc.)")
print("4. Check if a person can get a discount (age < 18 or > 60)")
```

```python
# ========== FOR LOOPS ==========
print("="*50)
print("FOR LOOPS")
print("="*50)

# Loop through a list
fruits = ["apple", "banana", "orange", "grape"]
print("Fruits list:")
for fruit in fruits:
    print(f"- {fruit}")

# Loop through a string
print("\nCharacters in 'Python':")
for char in "Python":
    print(f"'{char}'")

# Loop with range()
print("\nNumbers 1 to 5:")
for i in range(1, 6):
    print(i)

# Loop with step
print("\nEven numbers 0 to 10:")
for i in range(0, 11, 2):
    print(i)

# Nested loops
print("\nMultiplication table:")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}")
    print("-" * 10)

# Loop with enumerate (get index and value)
print("\nFruits with index:")
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

# Loop through dictionary
student = {"name": "Mita", "age": 21, "course": "CS"}
print("\nStudent info:")
for key, value in student.items():
    print(f"{key}: {value}")

# List comprehension (quick way to create lists)
print("\nList comprehension:")
squares = [x**2 for x in range(1, 6)]
print(f"Squares: {squares}")

even_numbers = [x for x in range(1, 11) if x % 2 == 0]
print(f"Even numbers: {even_numbers}")

# ========== EXERCISES ==========
print("\n" + "="*50)
print("YOUR TURN!")
print("="*50)
print("1. Print all numbers from 1 to 20")
print("2. Print only odd numbers from 1 to 20")
print("3. Sum all numbers from 1 to 100")
print("4. Create a list of squares of numbers 1-10")
print("5. Print a pattern using nested loops:")
print("   *")
print("   **")
print("   ***")
print("   ****")
```

```python
# ========== WHILE LOOPS ==========
print("="*50)
print("WHILE LOOPS")
print("="*50)

# Basic while loop
count = 1
print("Counting to 5:")
while count <= 5:
    print(count)
    count += 1

# While with user input (simulated)
print("\nGuessing game (simulated):")
secret = 7
guess = 0
attempts = 0

while guess != secret:
    attempts += 1
    guess = attempts  # Simulating guesses
    print(f"Attempt {attempts}: Guessed {guess}")
    if guess < secret:
        print("  Too low!")
    elif guess > secret:
        print("  Too high!")
    else:
        print(f"  Correct! Found in {attempts} attempts!")

# ========== BREAK & CONTINUE ==========
print("\n" + "="*50)
print("BREAK & CONTINUE")
print("="*50)

# Break - exit loop early
print("Break example (stop at 5):")
for i in range(1, 11):
    if i == 5:
        break
    print(i)

# Continue - skip iteration
print("\nContinue example (skip even numbers):")
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)

# ========== EXERCISES ==========
print("\n" + "="*50)
print("YOUR TURN!")
print("="*50)
print("1. Use while loop to print numbers 10 to 1")
print("2. Use while loop to sum numbers until sum > 100")
print("3. Print numbers 1-20, but skip numbers divisible by 3")
print("4. Print numbers 1-20, stop when you hit 13")
```

```python
# ========== FUNCTIONS ==========
print("="*50)
print("FUNCTIONS")
print("="*50)

# Basic function
def greet():
    print("Hello, World!")

greet()

# Function with parameters
def greet_person(name):
    print(f"Hello, {name}!")

greet_person("Mita")
greet_person("AI Student")

# Function with return value
def add(a, b):
    return a + b

result = add(5, 3)
print(f"5 + 3 = {result}")

# Function with multiple returns
def get_min_max(numbers):
    return min(numbers), max(numbers)

nums = [10, 5, 8, 20, 3]
minimum, maximum = get_min_max(nums)
print(f"Min: {minimum}, Max: {maximum}")

# Default parameters
def power(base, exponent=2):
    return base ** exponent

print(f"Square of 5: {power(5)}")
print(f"Cube of 5: {power(5, 3)}")

# Variable number of arguments
def sum_all(*numbers):
    return sum(numbers)

print(f"Sum: {sum_all(1, 2, 3, 4, 5)}")
print(f"Sum: {sum_all(10, 20, 30)}")

# Keyword arguments
def describe_person(name, age, city="Unknown"):
    return f"{name} is {age} years old from {city}"

print(describe_person("Mita", 21, "Singapore"))
print(describe_person("John", 25))  # Uses default city

# Lambda functions (small anonymous functions)
square = lambda x: x ** 2
print(f"Square of 5: {square(5)}")

# ========== EXERCISES ==========
print("\n" + "="*50)
print("YOUR TURN!")
print("="*50)
print("1. Create a function that checks if a number is even")
print("2. Create a function that calculates factorial")
print("3. Create a function that checks if a string is palindrome")
print("4. Create a function that returns average of a list of numbers")
print("5. Create a function with default parameters")
```
