```python
import numpy as np
import matplotlib.pyplot as plt

print("="*60)
print("📐 LINEAR ALGEBRA: VECTORS")
print("="*60)

# ========== 1. CREATING VECTORS ==========
print("\n1. CREATING VECTORS")

# Create vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print(f"Vector v1: {v1}")
print(f"Vector v2: {v2}")
print(f"v1 shape: {v1.shape}")

# ========== 2. VECTOR OPERATIONS ==========
print("\n2. VECTOR OPERATIONS")

# Addition
print(f"v1 + v2 = {v1 + v2}")

# Subtraction
print(f"v1 - v2 = {v1 - v2}")

# Scalar multiplication
print(f"3 * v1 = {3 * v1}")

# ========== 3. DOT PRODUCT ==========
print("\n3. DOT PRODUCT")

dot_product = np.dot(v1, v2)
print(f"v1 · v2 = {dot_product}")
print(f"Formula: 1*4 + 2*5 + 3*6 = {1*4 + 2*5 + 3*6}")

# ========== 4. VECTOR MAGNITUDE ==========
print("\n4. VECTOR MAGNITUDE")

magnitude_v1 = np.linalg.norm(v1)
magnitude_v2 = np.linalg.norm(v2)
print(f"|v1| = {magnitude_v1:.2f}")
print(f"|v2| = {magnitude_v2:.2f}")

# ========== 5. UNIT VECTOR ==========
print("\n5. UNIT VECTOR")

unit_v1 = v1 / magnitude_v1
print(f"Unit vector of v1: {unit_v1}")
print(f"Length of unit vector: {np.linalg.norm(unit_v1):.2f}")

# ========== 6. ANGLE BETWEEN VECTORS ==========
print("\n6. ANGLE BETWEEN VECTORS")

cos_angle = dot_product / (magnitude_v1 * magnitude_v2)
angle_rad = np.arccos(cos_angle)
angle_deg = np.degrees(angle_rad)
print(f"cos(θ) = {cos_angle:.4f}")
print(f"θ = {angle_deg:.2f} degrees")

# ========== 7. VISUALIZING VECTORS ==========
print("\n7. VISUALIZING VECTORS (2D)")

# Create 2D vectors for visualization
a = np.array([2, 3])
b = np.array([4, 1])

plt.figure(figsize=(8, 6))
# Plot vectors
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a = (2,3)')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='b = (4,1)')
plt.quiver(0, 0, a[0]+b[0], a[1]+b[1], angles='xy', scale_units='xy', scale=1, color='green', label='a+b = (6,4)')

plt.xlim(-1, 7)
plt.ylim(-1, 5)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)
plt.legend()
plt.title("Vector Addition Visualization")
plt.show()

print("\n🤖 WHY THIS MATTERS IN AI:")
print("- Data points are vectors (features)")
print("- Weights in neural networks are vectors")
print("- Dot product measures similarity")
print("- Gradient descent updates weight vectors")
```

```python
import numpy as np

print("="*60)
print("📐 LINEAR ALGEBRA: MATRICES")
print("="*60)

# ========== 1. CREATING MATRICES ==========
print("\n1. CREATING MATRICES")

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(f"Matrix A:\n{A}")
print(f"Matrix B:\n{B}")

# ========== 2. MATRIX OPERATIONS ==========
print("\n2. MATRIX OPERATIONS")

print(f"A + B:\n{A + B}")
print(f"A - B:\n{A - B}")
print(f"3 * A:\n{3 * A}")

# ========== 3. MATRIX MULTIPLICATION ==========
print("\n3. MATRIX MULTIPLICATION")

C = np.dot(A, B)
print(f"A @ B:\n{C}")

# ========== 4. TRANSPOSE ==========
print("\n4. TRANSPOSE")

print(f"A^T:\n{A.T}")

# ========== 5. IDENTITY MATRIX ==========
print("\n5. IDENTITY MATRIX")

I = np.eye(2)
print(f"I:\n{I}")

# ========== 6. INVERSE ==========
print("\n6. INVERSE")

inv_A = np.linalg.inv(A)
print(f"A^-1:\n{inv_A}")

# Verify: A * A^-1 = I
print(f"A * A^-1:\n{np.dot(A, inv_A)}")

# ========== 7. SOLVING LINEAR SYSTEMS ==========
print("\n7. SOLVING LINEAR SYSTEMS")
print("Solve: 2x + 3y = 8, x + y = 3")

# Matrix form: A * [x, y] = b
A = np.array([[2, 3], [1, 1]])
b = np.array([8, 3])

solution = np.linalg.solve(A, b)
print(f"x = {solution[0]:.2f}, y = {solution[1]:.2f}")

print("\n🤖 WHY THIS MATTERS IN AI:")
print("- Neural networks are matrix multiplications")
print("- Weight matrices store model parameters")
print("- Data batches are matrices")
print("- Solving linear systems = regression")
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

print("="*60)
print("📐 CALCULUS: DERIVATIVES")
print("="*60)

# ========== 1. SYMBOLIC DIFFERENTIATION ==========
print("\n1. SYMBOLIC DIFFERENTIATION")

x = sp.Symbol('x')

# Function: f(x) = x^2 + 3x + 1
f = x**2 + 3*x + 1
df = sp.diff(f, x)

print(f"f(x) = {f}")
print(f"f'(x) = {df}")

# ========== 2. PARTIAL DERIVATIVES ==========
print("\n2. PARTIAL DERIVATIVES")

x, y = sp.symbols('x y')
f = x**2 + y**2 + 3*x*y

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print(f"f(x,y) = {f}")
print(f"∂f/∂x = {df_dx}")
print(f"∂f/∂y = {df_dy}")

# Gradient
gradient = [df_dx, df_dy]
print(f"Gradient = {gradient}")

# ========== 3. CHAIN RULE ==========
print("\n3. CHAIN RULE")

# f(x) = sin(x^2 + 3x)
inner = x**2 + 3*x
outer = sp.sin(inner)
derivative = sp.diff(outer, x)

print(f"f(x) = sin(x² + 3x)")
print(f"f'(x) = {derivative}")

# ========== 4. GRADIENT DESCENT VISUALIZATION ==========
print("\n4. GRADIENT DESCENT")

# Find minimum of f(x) = x^2 + 3x + 1
def f(x):
    return x**2 + 3*x + 1

def df(x):
    return 2*x + 3

# Gradient descent
x_start = 10
learning_rate = 0.1
iterations = 20

x_path = [x_start]
for _ in range(iterations):
    x_current = x_path[-1]
    x_next = x_current - learning_rate * df(x_current)
    x_path.append(x_next)

# Plot
x_vals = np.linspace(-8, 12, 100)
y_vals = f(x_vals)
y_path = [f(x) for x in x_path]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = x² + 3x + 1')
plt.scatter(x_path, y_path, color='red', zorder=5, label='Gradient descent path')
plt.plot(x_path, y_path, 'r--', alpha=0.5)
plt.axvline(x=-1.5, color='green', linestyle='--', label='Minimum at x = -1.5')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Finding the Minimum')
plt.show()

print(f"Starting at x = {x_start}")
print(f"Final x = {x_path[-1]:.4f}")
print(f"Expected minimum at x = -1.5")

print("\n🤖 WHY THIS MATTERS IN AI:")
print("- Gradient descent trains neural networks")
print("- Derivatives tell us direction to move")
print("- Chain rule = backpropagation")
print("- Partial derivatives optimize multiple parameters")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

print("="*60)
print("📊 PROBABILITY & STATISTICS")
print("="*60)

# ========== 1. BASIC PROBABILITY ==========
print("\n1. BASIC PROBABILITY")

# Coin toss
total = 2
favorable = 1  # Heads
p_head = favorable / total
print(f"P(Heads) = {p_head}")

# Die roll
total = 6
favorable = 3  # Even: 2,4,6
p_even = favorable / total
print(f"P(Even) = {p_even}")

# ========== 2. CONDITIONAL PROBABILITY ==========
print("\n2. CONDITIONAL PROBABILITY")

# Example: Spam detection
P_spam = 0.2
P_not_spam = 0.8
P_viagra_given_spam = 0.8
P_viagra_given_not_spam = 0.05

P_viagra = P_viagra_given_spam * P_spam + P_viagra_given_not_spam * P_not_spam
P_spam_given_viagra = (P_viagra_given_spam * P_spam) / P_viagra

print(f"P(Spam | Contains 'Viagra') = {P_spam_given_viagra:.3f}")
print(f"→ {P_spam_given_viagra*100:.1f}% likely to be spam")

# ========== 3. DESCRIPTIVE STATISTICS ==========
print("\n3. DESCRIPTIVE STATISTICS")

# Generate sample data
data = np.random.normal(50, 10, 1000)

print(f"Mean: {np.mean(data):.2f}")
print(f"Median: {np.median(data):.2f}")
print(f"Std Dev: {np.std(data):.2f}")
print(f"Variance: {np.var(data):.2f}")

# Quartiles
print(f"Q1: {np.percentile(data, 25):.2f}")
print(f"Q2 (Median): {np.percentile(data, 50):.2f}")
print(f"Q3: {np.percentile(data, 75):.2f}")

# ========== 4. DISTRIBUTIONS ==========
print("\n4. DISTRIBUTIONS")

# Normal distribution
x = np.linspace(-4, 4, 100)
normal = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(x, normal)
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')

# Uniform distribution
uniform = np.ones_like(x) / 8
plt.subplot(1, 3, 2)
plt.plot(x, uniform)
plt.title('Uniform Distribution')
plt.xlabel('x')

# Histogram of sample data
plt.subplot(1, 3, 3)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Sample Data Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# ========== 5. BAYES THEOREM - SPAM FILTER ==========
print("\n5. BUILDING A SIMPLE SPAM FILTER")

# Training data
words = {
    'spam': {'viagra': 0.8, 'free': 0.7, 'money': 0.6, 'hello': 0.1},
    'not_spam': {'viagra': 0.05, 'free': 0.1, 'money': 0.1, 'hello': 0.6}
}

P_spam = 0.2
P_not_spam = 0.8

def classify_email(text):
    """Simple Naive Bayes spam classifier"""
    words_in_email = text.lower().split()
    
    # Calculate P(Spam | words)
    P_spam_given_words = P_spam
    for word in words_in_email:
        if word in words['spam']:
            P_spam_given_words *= words['spam'][word]
    
    # Calculate P(Not Spam | words)
    P_not_spam_given_words = P_not_spam
    for word in words_in_email:
        if word in words['not_spam']:
            P_not_spam_given_words *= words['not_spam'][word]
    
    # Normalize
    total = P_spam_given_words + P_not_spam_given_words
    P_spam_final = P_spam_given_words / total
    
    return P_spam_final

# Test emails
email1 = "viagra free money"
email2 = "hello friend how are you"

p1 = classify_email(email1)
p2 = classify_email(email2)

print(f"Email: '{email1}' → {p1*100:.1f}% chance of spam")
print(f"Email: '{email2}' → {p2*100:.1f}% chance of spam")

print("\n🤖 WHY THIS MATTERS IN AI:")
print("- Probability powers uncertainty in AI")
print("- Bayes theorem = Naive Bayes classifier")
print("- Distributions = modeling data")
print("- Statistics = evaluating models")
```
